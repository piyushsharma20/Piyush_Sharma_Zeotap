```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
customers = pd.read_csv('/content/Customers.csv')
```

```python
products= pd.read_csv('/content/Products.csv')
```

```python
transactions = pd.read_csv('/content/Transactions.csv')
```

```python
merged_data = pd.merge(transactions, customers, on='CustomerID', how='left')
```

```python
merged_data.head()
```

```python
merged_data.info()
```

```python
merged_data.isnull().sum()
```

```python
merged_data = pd.merge(merged_data, products, on='ProductID', how='left')
```

```python
merged_data.head()
```

```python
merged_data = merged_data.drop(columns=['Price_y'])
merged_data = merged_data.rename(columns={'Price_x': 'Price'})
```

```python
merged_data.head()
```

```python
print("\nStatistical Summary:")
print(merged_data.describe())
```

```python
categorical_cols = ['CustomerID', 'ProductID', 'CustomerName', 'Region', 'ProductName', 'Category']
for col in categorical_cols:
    print(f"\nUnique values in {col}:")
    print(merged_data[col].nunique())
```

```python
merged_data['TransactionDate'] = pd.to_datetime(merged_data['TransactionDate'])
merged_data['SignupDate'] = pd.to_datetime(merged_data['SignupDate'])
```

```python
merged_data['TransactionMonthYear'] = merged_data['TransactionDate'].dt.to_period('M')
```

```python
merged_data.head()
```

```python
plt.figure(figsize=(12, 6))
```

```python
# Correlation Matrix for numerical features
correlation_matrix = merged_data[['Quantity', 'TotalValue', 'Price']].corr()
print(correlation_matrix)
```

```python
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', cbar=True)
plt.title('Correlation Matrix')
plt.show()
```

```python
sns.boxplot(x=merged_data['TotalValue'], color='lightgreen')
plt.title('Boxplot of Total Value')
plt.show()
```

```python
sns.boxplot(x=merged_data['Price'], color='blue')
plt.title('Boxplot of Total Value')
plt.show()
```

```python
sns.histplot(merged_data['TotalValue'], bins=30, kde=True, color='blue')
plt.title('Distribution of Total Transaction Values')
plt.xlabel('Total Value (USD)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(data=merged_data, x='Region', palette='Set2', order=merged_data['Region'].value_counts().index)
plt.title('Number of Transactions by Region')
plt.xlabel('Region')
plt.ylabel('Number of Transactions')
plt.show()
```

```python
top_products = merged_data['ProductName'].value_counts().head(10)
top_products.plot(kind='bar', figsize=(10, 5), color='green')
plt.title('Top 10 Most Purchased Products')
plt.xlabel('Product Name')
plt.ylabel('Number of Purchases')
plt.xticks(rotation=45)
plt.show()
```

```python
merged_data.groupby('TransactionMonthYear')['TransactionID'].count().plot(kind='line', marker='o', figsize=(12, 6), color='orange')
plt.title('Monthly Transaction Trend')
plt.xlabel('Month-Year')
plt.ylabel('Number of Transactions')
plt.grid()
plt.show()
```

```python
category_revenue = merged_data.groupby('Category')['TotalValue'].sum().sort_values(ascending=False)
category_revenue.plot(kind='bar', figsize=(10, 5), color='purple')
plt.title('Total Revenue by Product Category')
plt.xlabel('Category')
plt.ylabel('Total Revenue (USD)')
plt.xticks(rotation=45)
plt.show()
```

```python
monthly_category = merged_data.groupby(['TransactionMonthYear', 'Category'])['TotalValue'].sum().unstack()
monthly_category.plot(kind='bar', stacked=True, figsize=(15, 7), colormap='viridis')
plt.title('Monthly Revenue by Product Category')
plt.xlabel('Month-Year')
plt.ylabel('Total Revenue (USD)')
plt.legend(title='Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
region_category = merged_data.groupby(['Region', 'Category'])['TotalValue'].sum().unstack()
region_category.plot(kind='bar', stacked=True, figsize=(15, 7), colormap='coolwarm')
plt.title('Transaction Cost by Category in Each Region')
plt.xlabel('Region')
plt.ylabel('Total Transaction Cost (USD)')
plt.legend(title='Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

max_revenue_category = region_category.idxmax(axis=1)
print("\nMaximum Revenue Generating Category in Each Region:")
print(max_revenue_category)
```

```python
monthly_region_sales = merged_data.groupby(['TransactionMonthYear', 'Region'])['TotalValue'].sum().unstack()
monthly_region_sales.plot(kind='line', marker='o', figsize=(15, 7), colormap='plasma')
plt.title('Month-wise Sales in Each Region')
plt.xlabel('Month-Year')
plt.ylabel('Total Sales (USD)')
plt.legend(title='Region', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
top_least_products = merged_data.groupby(['Region', 'ProductName'])['Quantity'].sum().reset_index()
top_least_products_sorted = top_least_products.sort_values(by=['Region', 'Quantity'], ascending=[True, False])
top_3_products = top_least_products_sorted.groupby('Region').head(3)
least_3_products = top_least_products_sorted.groupby('Region').tail(3)

print("\nTop 3 Products in Each Region:")
print(top_3_products[['Region', 'ProductName']])

print("\nLeast 3 Products in Each Region:")
print(least_3_products[['Region', 'ProductName']])
```
